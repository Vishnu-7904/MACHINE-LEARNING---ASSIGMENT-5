```python
#importing all the required modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import math
import seaborn as sns
from scipy.stats import randint as sp_randInt
from sklearn.model_selection import RandomizedSearchCV

df = pd.read_excel("C:/Users/91914/Downloads/weatherAUS.xlsx") #Loading our Dataset
df=pd.DataFrame(df)

def removeNull(Class): # Funtion to merge 2 features in one array
    iterate=df[Class]
    y=[]
    mean=df[Class].mean(axis=0)
    for i in iterate:
        if(math.isnan(i)):
            y.append(mean)
        else:
            y.append(i)
    y=pd.DataFrame(y)
    df[Class]=y

# Removing the Null value and replacing it with the mean of the column
removeNull('WindGustSpeed')
removeNull('WindSpeed9am')

def isNaN(string):
    return string != string

# Replacing null values with the most frequent value in the column.
def removeNullCategorical(Class):
    iterate=df[Class]
    y=[]
    mostfreq=df[Class].value_counts().idxmax()
    for i in iterate:
        if(isNaN(i)):
            y.append(mostfreq)
        else:
            y.append(i)
    y=pd.DataFrame(y)
    df[Class]=y
    return y
removeNullCategorical('RainToday')

# Extracting features and target variables
y=df['RainToday'].to_numpy()
y=y[:600]
X=df[['WindGustSpeed','WindSpeed9am']].to_numpy()
X=X[:600]

# Splitting the data into train and set
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.5)

# Applying KNN of K=3
neigh=KNeighborsClassifier(n_neighbors=3)

# Training the model
neigh.fit(X_train,y_train)

# Returning the prediction vector predicted by the model
test_vector=neigh.predict(X_test)

# Returning the confusion matrix and the report
print(confusion_matrix(y_test, test_vector))
print(classification_report(y_test,test_vector))

# Returning the prediction vector predicted by the model
test_vector=neigh.predict(X_train)

# Returning the confusion matrix and the report
print(classification_report(y_train,test_vector))

# Returning the accuracy of the model with respect to testing model
print('testing_accuracy:',neigh.score(X_test,y_test))
print('training_accuracy:',neigh.score(X_train,y_train))
test_accuracy=neigh.score(X_test,y_test)
train_accuracy=neigh.score(X_train,y_train)
print(f"lets consider 80% as a good accuracy")

# Function to check the fitting of the model
def checkfitting(test_accuracy,train_accuracy):
    if(train_accuracy < 0.80):
        return ("It is Under-fitting")
    elif(train_accuracy > 0.80 and test_accuracy > 0.80):
        return ("It is Good-fitting")
    elif(test_accuracy < 0.80):
        return ("It is Over-fitting")
print(checkfitting(test_accuracy,train_accuracy))

# LAB 3 part
# Convert the excel contents to dataframe
df = pd.read_excel("C:/Users/91914/Downloads/Lab Session1 Data (1).xlsx", sheet_name='Purchase data')
df=pd.DataFrame(df)

# Convert dataframe to array
Array=df.to_numpy()
A=Array[:,1:4]
C=Array[:,4:5]
A=np.float64(A)

# Finding the pseudoinverse
pinv=np.linalg.pinv(A)
print("pseudo inverse:",pinv)

# Calculating the cost of each product
print("Costs of the products:",np.matmul(pinv,C))
pinv2=np.matmul(pinv,C)

# Comparing real prices with the model for calculating prices
print("real prices:",C)
predicted=np.matmul(A,pinv2)
print("model of calculating prices:",predicted)

# Function to calculate MSE
def MSE(C,predicted):
    sum=0
    for i in range(len(C)):
        sub=0
        sub=C[i]-predicted[i]
        sub=sub*sub
        sum+=sub
    return sum/len(C)
print("Mean squared error:",MSE(C,predicted))

# Function to calculate RMSE
def RMSE(C,predicted):
    sum=0
    for i in range(len(C)):
        sub=0
        sub=C[i]-predicted[i]
        sub=sub*sub
        sum+=sub
    return math.sqrt(sum/len(C))
print("RMSE:",RMSE(C,predicted))

# Function to calculate MAPE
def MAPE(C,predicted):
    sum=0
    for i in range(len(C)):
        sub=0
        sub=C[i]-predicted[i]
        sub=abs((sub)/C[i])
        sum+=sub
    return sum/len(C)
print("MAPE:",MAPE(C,predicted))

# Function to calculate R2 score
def R2(C,predicted):
    rss=0
    for i in range(len(C)):
        err=0
        err=C[i]-predicted[i]
        rss += err*err
    tss=0
    mean=np.mean(predicted)
    for i in range(len(C)):
        err=0
        err=C[i]-mean
        tss += err*err
    return 1-(rss/tss)
print("R2",R2(C,predicted))

def classify(k):
    x=np.random.randint(1,11,size=20)
    y=np.random.randint(1,11,size=20)

# Initialzing the data co-ordinates
    c0x=[]
    c0y=[]
    c1x=[]
    c1y=[]
    class0=[]
    class1=[]

# Sum of the datapoints
    target=[]
    for i in range(20):
        if((x[i]+y[i])% 2==0):
            c0x.append(x[i])
            c0y.append(y[i])
            target.append("blue")
        else:
            c1x.append(x[i])
            c1y.append(y[i])
            target.append("red")
    class0.append(c0x)
    class0.append(c0y)
    class1.append(c1x)
    class1.append(c1y)

    x=np.array(x)
    y=np.array(y)
    X=[]
    X.append(x)
    X.append(y)
    X=np.array(X)
    X=X.transpose()
    y_train=target

# Plotting the class1 and class2
    plt.scatter(class1[0],class1[1],color='red',label='class1')
    plt.scatter(class0[0],class0[1],color='blue',label='class0')

    x=np.arange(0,10,0.1)
    y=np.arange(0,10,0.1)

    X_test=[]
    X_test.append(x)
    X_test.append(y)
    X_test=np.array(X_test)
    X_test=X_test.transpose()

    neigh=KNeighborsClassifier(n_neighbors=k)
    neigh.fit(X,y_train)
    test_points=neigh.predict(X_test)
    test_points=np.array(test_points)
    c0x=[]
    c0y=[]
    c1x=[]
    c1y=[]
    class0=[]
    class1=[]
    for i in range(len(test_points)):
        if(test_points[i]=="blue"):
            c0x.append(x[i])
            c0y.append(y[i])
        else:
            c1x.append(x[i])
            c1y.append(y[i])
    class0.append(c0x)
    class0.append(c0y)
    class1.append(c1x)
    class1.append(c1y)
    plt.scatter(class1[0],class1[1],color='red')
    plt.scatter(class0[0],class0[1],color='blue')
    plt.show()
classify(3)
classify(2)
classify(4)

def classifyData(X,y,k):

    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.5) # Splitting the data into train and set

# Applying KNN of K=3
    neigh=KNeighborsClassifier(n_neighbors=k)
    x=X_train[:,0]
    y=X_train[:,1]
    x=x.transpose()
    y=y.transpose()
    neigh.fit(X_train,y_train) # Training the model

# Returning the prediction vector predicted by the model
    test_vector=neigh.predict(X_test)
    f2=plt.figure(1)
    sns.scatterplot(x=X_train[:,0], y=X_train[:,1], hue=y_train, palette=['blue','red'])
    f2=plt.figure(2)
    sns.scatterplot(x=X_test[:,0], y=X_test[:,1], hue=y_test, palette=['blue','red'])
    plt.show()
classifyData(X,y,3)
classifyData(X,y,4)
classifyData(X,y,2)


X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.5) # Splitting the data into train and set
neigh=KNeighborsClassifier(n_neighbors=3) # Applying KNN of K=3
parameters={'n_neighbors':sp_randInt(1,100)} # Initialzing n_neighbors

# Calling randomized Search CV to tune the hyperparameter
clf = RandomizedSearchCV(neigh, param_distributions=parameters,random_state=0)
search=clf.fit(X,y)
print("best K value:",search.best_params_)
```
